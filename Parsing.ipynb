```python
headers = {
 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate',
    'Connection': 'keep-alive'
}
def get_url():
    for count in range(1, 30):
        sleep(0)   
        URL = f"https://stopgame.ru/review/p{count}?subsection=izumitelno"
        response = requests.get(URL, headers=headers)
        soup = bs(response.text, 'lxml')
        data = soup.find_all('div', attrs={"data-key": re.compile(r'^\d+$')}) #attrs 
        for i in data:
            card_url = "https://stopgame.ru" + i.find('a').get("href")
            yield card_url
        

for card_url in get_url():
    response = requests.get(card_url, headers=headers)
    sleep(0)
    soup = bs(response.text, 'lxml')
    data = soup.find('section', class_='_page-section_1jnog_471 _section_12po9_6 _content-section_12po9_13')
    
    if data is not None:
       # Находим имя
        name_tag = data.find("h1", attrs={"style": "view-transition-name: material-title"})
        name = name_tag.text if name_tag else "Имя не найдено"
        
        # Находим изображение
        img_tag = data.find("div", class_="_image-wrapper_12po9_173 _image-width_12po9_116")
        img = img_tag.find("a").get("href") if img_tag and img_tag.find("a") else "Изображение не найдено"
      
        
        # Проверяем наличие видео
        video_img = soup.find('lite-youtube')
        if video_img is not None:
            video_style = video_img.get("style")
            url = video_style.split('url(')[-1].rstrip(')') if video_style else "URL видео не найден"
            print(name, url, sep="\n")
        else: 
            print(name, img, sep="\n")
    else:
        print(f"Раздел не найден для URL: {card_url}")
```
